# Qubit
***

A classical computer stores information in bits. Bit can be either 0 or 1 <em>(True or False)</em>. Where as a Quantum bit (the smallest unit for Quantum Computation) can be in any state which is a linear combination of $|0\rangle$ and $|1\rangle$.

$$
    |\psi\rangle = a|0\rangle + b|1\rangle\text{ with }a,b \in \mathbb{C} \text{ and } |a|^2 + |b|^2 = 1
$$

a and b being complex numbers we have four real degrees of freedoms. Overall constant phase does not matter which reduces one of the dof whereas the constraint $|a|^2 + |b|^2 = 1$ removes another one. Sowe are left with two real degrees of freedoms. One can conviniently represent this states of qubit in the following form

$$
    |\psi\rangle = e^{i \eta} \left(cos\left(\frac{\theta}{2}\right)|0\rangle + e^{i \phi}sin\left(\frac{\theta}{2}\right)|1\rangle \right)
$$

Overall phase $e^{i \eta}$ is irrelevant and cannot be measured physically. So any qubit state can be parameterised by two real variable $\theta\in [0,\pi]$ and $\phi\in[0,2\pi)$. Then states can be identified with the points on a sphere ($S^2$) called the Bloch sphere. 

```python
from qiskit import *
from qiskit.visualization import plot_bloch_vector
```

```python
vector = [0,0,1] ## [x,y,z] = [sin(theta)cos(phi), sin(theta)sin(phi), cos(theta)]
plot_bloch_vector(vector)
```

Note that the states $|0\rangle$ and $|1\rangle$ are at antipodal positions.
